```python
# 1. Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# 2. Read the dataset
df = pd.read_csv("insurance_data.csv")

# 3. Display first 5 rows
print(df.head())

# 4. Scatter plot to visualize data
plt.scatter(df.age, df.bought_insurance, marker='+', color='red')
plt.xlabel("Age")
plt.ylabel("Bought Insurance")
plt.title("Age vs Bought Insurance")
plt.show()

# 5. Split data into input (X) and output (y)
X = df[['age']]              # Independent variable
y = df['bought_insurance']   # Dependent variable

# 6. Train-test split (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.8, random_state=42
)

# 7. Create Logistic Regression model
model = LogisticRegression()

# 8. Train the model
model.fit(X_train, y_train)

# 9. Predict using test data
y_predicted = model.predict(X_test)

# 10. Print predictions
print("Predicted values:", y_predicted.tolist())
print("Actual values   :", y_test.tolist())

# 11. Predict probabilities
probabilities = model.predict_proba(X_test)
print("\nPrediction Probabilities:")
print(probabilities)

# 12. Check model accuracy
accuracy = model.score(X_test, y_test)
print("\nModel Accuracy:", accuracy)

# 13. Predict for a new age value
age = [[35]]
prediction = model.predict(age)
prob = model.predict_proba(age)

print("\nPrediction for age 35:")
print("Will buy insurance?" , prediction[0])
print("Probability:", prob)
```
